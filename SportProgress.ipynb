```python
'''
Plot results of sport progress.
'''
%matplotlib inline
from datetime import datetime
from pprint import pprint

def get_data(filename):
    items = dict()
    with open(filename) as f:
        for i, line in enumerate(f):
            idx = line.index(":")
            day = datetime.strptime(line[:idx].strip() + " 2015", "%d %b %Y")
            activity = line[idx+2:].split()[0]
            info_idx = idx+3+len(activity)
            info = line[info_idx:]
            items[i] = (day, activity, info)
    return items
        
items = get_data("January.txt")
```

```python
def get_weight(items):
    day2weight = dict()
    for (day, _, info) in items.values():
        # if there is a weight info in the line
        if info.strip() and ")" == info.strip()[-1]:
            weight_idx1 = info[-15:].index("(")
            weight_idx2 = info[-15:].index("lb")
            weight = info[len(info) - 15 + weight_idx1 + 1: len(info) - 15 + weight_idx2-1]
            weight = float(weight)
            # if we have two weights per day, average them
            if day in day2weight:
                day2weight[day] = (day2weight[day] + weight)/2
            else:
                day2weight[day] = weight
    return day2weight

day2weight = get_weight(items)
```

```python
import prettyplotlib as ppl
import matplotlib.pyplot as plt
import matplotlib

x = []
y = []
for (d, w) in sorted(day2weight.items()):
    x.append(d)
    y.append(w)

fig = plt.figure(figsize=(18,6))
ax = fig.add_subplot(111)
ax.plot_date(x, y, 'r-o')
ax.set_xticks(x)
ax.set_xticklabels(
        [date.strftime("%d %b") for (date, value) in sorted(day2weight.items())]
        )
fig.autofmt_xdate()

plt.ylim([165, 175])
plt.xlabel('Date')
plt.ylabel('Weight (lb.)')
plt.title('Weight Changes. January.', fontsize = 18)
plt.show()
fig.savefig('weight_january.png')
```
